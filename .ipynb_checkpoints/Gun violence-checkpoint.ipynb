```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
engine = create_engine("postgresql+psycopg2://postgres:postgres@localhost:5433/Gunviolence_db")
```

```python
engine.connect()
```

```python
engine.table_names()
```

```python
#check if accidental death table sql queries work

accidental_death_killed = pd.read_sql_query('select * from accidental_deaths', con=engine)
accidental_death_killed = accidental_death_killed.groupby(['state']).sum()
accidental_death_killed.head()
```

```python
#check mass shooting table sql connection work 

mass_shooting_killed_df =pd.read_sql_query('select * from mass_shooting', con=engine)
mass_shooting_killed_df = mass_shooting_killed_df.groupby(['state']).sum()
mass_shooting_killed_df.head()
```

```python
#check if sql school shootings table queries work 

school_shooting_killed = pd.read_sql_query('select * from school_shootings', con=engine)
school_shooting_killed = school_shooting_killed.groupby(['state']).sum()
school_shooting_killed.head()
```

```python
#Join all three table
gun_violence1 = pd.merge(accidental_death_killed, mass_shooting_killed_df, on = "state", how = "inner")
gun_violence1.head()
```

```python
gun_violence = pd.merge(gun_violence1, school_shooting_killed, on = "state", how = "inner")
gun_violence.head(10)
```

```python
test = gun_violence
gun_violence.reset_index(inplace=True)
gun_violence.head()
```

```python
gun_crime = gun_violence.drop(['index'], axis=1)
gun_crime.head()
```

```python
#Load clean join dataframes back into SQL
gun_crime.to_sql(name='gun_crime', con=engine, if_exists='append', index=False)
```
